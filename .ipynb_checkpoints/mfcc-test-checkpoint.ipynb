```python
def write_data(data, path):
    directory = os.path.dirname(path)
    with suppress(FileExistsError):
        os.makedirs(directory)
    np.save(path, data)
    

audio_files_df = pd.DataFrame({'raw_path': [], 'output_path': [], 'label': []})

for root, directory, filenames in os.walk(audio_dir):
    for filename in filenames:
        if filename.endswith(".flac"):
            raw_path = os.path.join(root, filename)
            output_path = os.path.join(output_dir, root.split('\\')[-1], filename.replace('.flac', '.npy'))
            audio_files_df = audio_files_df.append({'raw_path': raw_path, 
                                                    'output_path': output_path, 
                                                    'label': root.split('\\')[-1]},
                                  ignore_index=True)

audio_files_df.shape
```

```python
def show_audio(signal):
    plt.figure(figsize=(12, 4))
    plt.plot(signal)
    plt.show()
```

```python
class AudioDataTransformer():
    def __init__(self, hp):
        # Keep track of hyper parameters and make the mel basis
        self.hp = hp
        self.make_mel_basis()
        
    def preemphasis(self, data):
        return scipy.signal.lfilter([1, -self.hp.preemphasis], [1,], data)

    def inv_preemphasis(self, data):
        return scipy.signal.lfilter([1,], [1, -self.hp.preemphasis], data)

    def short_time_ft(self, data):
        return librosa.stft(data, n_fft=self.hp.n_fft, 
                            hop_length=self.hp.hop_length, 
                            win_length=self.hp.win_length)

    def linear_to_mel(self, spectrogram):
        return np.dot(self.mel_basis, spectrogram)
    
    def make_mel_basis(self):
         self.mel_basis = librosa.filters.mel(self.hp.sample_rate, 
                                   self.hp.n_fft, 
                                   n_mels=self.hp.n_mels)
            
    def amp_to_db(self, x):
        return 20 * np.log10(np.maximum(1e-5, x))
    
    def normalize(self, x):
        mean = x.mean(axis=0)
        return (x - mean) / mean
    
    def create_spectrogram(self, signal, normalize=True):
        emphasized_signal = self.preemphasis(signal)
        amp_spectrogram = self.short_time_ft(emphasized_signal)
        db_spectrogram = self.amp_to_db(np.abs(amp_spectrogram))
        
        if normalize:
            db_spectrogram = self.normalize(db_spectrogram)
            
        return db_spectrogram
        
    def create_mel_spectrogram(self, signal, normalize=True):
        emphasized_signal = self.preemphasis(signal)
        amp_spectrogram = self.short_time_ft(emphasized_signal)
        amp_mel_spectrogram = self.linear_to_mel(np.abs(amp_spectrogram))
        db_mel_spectrogram = self.amp_to_db(amp_mel_spectrogram)
        
        if normalize:
            db_mel_spectrogram = self.normalize(db_mel_spectrogram)

        return db_mel_spectrogram
    
    def display_spectrogram(self, spectrogram):
        x, y = np.mgrid[:spectrogram.shape[1], :spectrogram.shape[0]]
        plt.pcolormesh(x * self.hp.hop_length / self.hp.sample_rate, y, spectrogram.T)
        plt.xlabel('Time (s)')
        plt.show()
```
